```python
import pandas as pd
price_demand_data = pd.read_csv('price_demand_data.csv')
price_demand_data["PRICECATEGORY"] = price_demand_data["PRICECATEGORY"].replace(["LOW", "MEDIUM", "HIGH", "EXTREME"],[1, 2, 3, 4])
price_demand_data[["DATE", "Time"]] = price_demand_data["SETTLEMENTDATE"].str.split(" ", expand=True)
price_demand_data = price_demand_data.groupby('DATE').agg({'TOTALDEMAND': 'max','PRICECATEGORY': 'max'})
price_demand_data["PRICECATEGORY"] = price_demand_data["PRICECATEGORY"].replace([1, 2, 3, 4],["LOW", "MEDIUM", "HIGH", "EXTREME"])
price_demand_data
```

```python
price_demand_data.to_csv('cleaned_price.csv', encoding='utf-8', index=False)
```
