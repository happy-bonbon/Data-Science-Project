```python
# linear regression = demand prediction model
import pandas as pd
import numpy as np
```

```python
Combined = pd.read_csv('1.Combined_detail_cleaned.csv',encoding= 'ISO-8859-1')
Combined.head()
```

```python
FEATURES = ['TOTALDEMAND',
            'Minimum temperature (Ã\x82Â°C)', 
            'Maximum temperature (Ã\x82Â°C)', 
            'Rainfall (mm)', 
            'Evaporation (mm)', 
            'Sunshine (hours)', 
            'Speed of maximum wind gust (km/h)',  
            '9am Temperature (Ã\x82Â°C)', 
            '9am relative humidity (%)',
            '9am cloud amount (oktas)',
            '9am wind speed (km/h)',
            '9am MSL pressure (hPa)',
            '3pm Temperature (Ã\x82Â°C)',
            '3pm relative humidity (%)',
            '3pm cloud amount (oktas)',
            '3pm wind speed (km/h)',
            '3pm MSL pressure (hPa)']
```

```python
Combined[FEATURES].corr(method='pearson')
```

```python
Select_FEATURES = [ 'Minimum temperature (Ã\x82Â°C)', 
                    'Maximum temperature (Ã\x82Â°C)', 
                    'Evaporation (mm)', 
                    'Sunshine (hours)', 
                    '9am Temperature (Ã\x82Â°C)', 
                    '9am relative humidity (%)',
                    '9am cloud amount (oktas)',
                    '9am wind speed (km/h)',
                    '3pm Temperature (Ã\x82Â°C)']
```

```python
X = Combined[Select_FEATURES]
y = Combined['TOTALDEMAND']
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import preprocessing
# Step 0: Preprocess
# randomly select 80% of the instances to be training and the rest to be testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
```

```python
# Step 1: Instantiate
lm = linear_model.LinearRegression()
```

```python
# Step 2: Fit
lm.fit(X_train, y_train)
```

```python
print(lm.coef_,lm.intercept_)
```

```python
r2_test = lm.score(X_test, y_test)
print(r2_test)
```

```python
# Step 3: Predict
y_pred = lm.predict(X_test)
```

```python
mean_squared_error(y_pred, y_test)
```

```python
r2_score(y_pred, y_test)
```

```python
# Step 4: Evaluate
print(f'mean squared error: {mean_squared_error(y_pred, y_test)}')
print(f'r2 score: {r2_score(y_pred, y_test)}')
```
